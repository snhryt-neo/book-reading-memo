# 1 序論

## 1.2 決定分析の役割

- 意思決定理論の役割として「意思決定という問題を表現してくれる」という側面がある
  - 決定分析、あるいはその基礎づけとなる意思決定理論で、人間の主観を一切排除して決定を下すのは難しい
  - 「意思決定理論はリスクについての決定を表現するための"言語"といえる」 (Fischhoff and Kadvany, 2015, p7)
- モデル
  - 本書では意思決定という操作を単純化、理想化したもののことを指す
  - モデルを明示的に用いることで、議論の対象を明確にできる
- 意思決定のアプローチ
  - **規範 (normative) 理論**: どのような意思決定が**望ましいか**を考える。いくつかの前提（公理）を置いて、その前提が満たされるときにとるべき行動について議論する
  - **記述 (descriptive) 理論**: どのような意思決定が**実際に行われているのか**を説明する
  - これらの知見を参考にして意思決定を支援するアプローチを処方的 (prescriptive) アプローチと呼ぶ → 決定分析はこれに該当
- 決定分析の役割: 数量化によって**意思決定の過程が明確**になり、**意思決定にかかわる人とのコミュニケーションが円滑**になる
  - 意思決定の過程が明確になれば、メタ的に「意思決定の過程の評価」も可能になる
